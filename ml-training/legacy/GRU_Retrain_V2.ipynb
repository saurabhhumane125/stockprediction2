```python
import numpy as np

LOOKBACK = 48

X = []
y = []

sequence_stock = []
sequence_date = []
sequence_df_index = []

for stock_id in sorted(df["Stock"].unique()):

    stock_df = df[df["Stock"] == stock_id].copy()

    feature_values = stock_df[FEATURES].values
    target_values = stock_df[TARGET].values

    stock_dates = stock_df["Date"].values
    stock_indices = stock_df.index.values

    for i in range(LOOKBACK, len(stock_df)):
        X.append(feature_values[i-LOOKBACK:i])
        y.append(target_values[i])

        sequence_stock.append(stock_id)
        sequence_date.append(stock_dates[i])
        sequence_df_index.append(stock_indices[i])

X = np.array(X, dtype=np.float32)
y = np.array(y, dtype=np.int64)

sequence_stock = np.array(sequence_stock)
sequence_date = np.array(sequence_date)
sequence_df_index = np.array(sequence_df_index)

print("X:", X.shape)
print("y:", y.shape)
print("Stock:", sequence_stock.shape)
print("Date:", sequence_date.shape)
print("Index:", sequence_df_index.shape)

print("\nFirst sequence")
print("Stock:", stock_encoder.inverse_transform([sequence_stock[0]])[0])
print("Date :", sequence_date[0])
print("Index:", sequence_df_index[0])
print("Target:", y[0])
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

train_idx = []
val_idx = []
test_idx = []

for stock_id in np.unique(sequence_stock):

    idx = np.where(sequence_stock == stock_id)[0]

    n = len(idx)

    train_end = int(n * 0.70)
    val_end = int(n * 0.85)

    train_idx.extend(idx[:train_end])
    val_idx.extend(idx[train_end:val_end])
    test_idx.extend(idx[val_end:])

train_idx = np.array(train_idx)
val_idx = np.array(val_idx)
test_idx = np.array(test_idx)

X_train = X[train_idx]
X_val = X[val_idx]
X_test = X[test_idx]

y_train = y[train_idx]
y_val = y[val_idx]
y_test = y[test_idx]

print("Train:", X_train.shape, y_train.shape)
print("Validation:", X_val.shape, y_val.shape)
print("Test:", X_test.shape, y_test.shape)

print("\nTotal sequences:",
      len(train_idx)+len(val_idx)+len(test_idx))
```

```python
from sklearn.preprocessing import MinMaxScaler

# Save original shapes
train_shape = X_train.shape
val_shape = X_val.shape
test_shape = X_test.shape

# Flatten to 2D
X_train_2d = X_train.reshape(-1, train_shape[-1])
X_val_2d = X_val.reshape(-1, val_shape[-1])
X_test_2d = X_test.reshape(-1, test_shape[-1])

# Fit ONLY on training data
scaler = MinMaxScaler()
scaler.fit(X_train_2d)

# Transform all sets
X_train = scaler.transform(X_train_2d).reshape(train_shape)
X_val = scaler.transform(X_val_2d).reshape(val_shape)
X_test = scaler.transform(X_test_2d).reshape(test_shape)

print("Train:", X_train.shape)
print("Validation:", X_val.shape)
print("Test:", X_test.shape)

print("\nTrain Min:", X_train.min())
print("Train Max:", X_train.max())
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Precision, Recall, AUC

model = Sequential([
    Input(shape=(48, 20)),

    GRU(
        units=64,
        return_sequences=True,
        recurrent_dropout=0.2
    ),
    Dropout(0.3),

    GRU(
        units=32,
        return_sequences=False,
        recurrent_dropout=0.2
    ),
    Dropout(0.3),

    Dense(16, activation="relu"),

    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        Precision(name="precision"),
        Recall(name="recall"),
        AUC(name="auc")
    ]
)

model.summary()
```

```python
from tensorflow.keras.callbacks import (
    EarlyStopping,
    ModelCheckpoint,
    ReduceLROnPlateau
)

early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True,
    verbose=1
)

checkpoint = ModelCheckpoint(
    "/content/best_gru_model.keras",
    monitor="val_loss",
    save_best_only=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)

callbacks = [
    early_stopping,
    checkpoint,
    reduce_lr
]

print("Callbacks created successfully.")
print(callbacks)
```

```python
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)
```

```python
SEQUENCE_LENGTH = 48

X = []
y = []
stocks = []
dates = []

for stock in df["Stock"].unique():

    stock_df = df[df["Stock"] == stock].reset_index(drop=True)

    stock_scaled = X_scaled[stock_df.index]

    for i in range(SEQUENCE_LENGTH, len(stock_df)):

        X.append(
            stock_scaled[i-SEQUENCE_LENGTH:i]
        )

        y.append(
            stock_df.loc[i, "Target"]
        )

        stocks.append(
            stock
        )

        dates.append(
            stock_df.loc[i, "Date"]
        )

X = np.array(X, dtype=np.float32)
y = np.array(y, dtype=np.int32)
stocks = np.array(stocks)
dates = np.array(dates)

print("X shape:", X.shape)
print("y shape:", y.shape)

print("\nSequence length:", X.shape[1])
print("Features:", X.shape[2])

print("\nFirst target:", y[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import AUC, Precision, Recall

model = Sequential([
    GRU(
        64,
        return_sequences=True,
        input_shape=(48, 20)
    ),

    Dropout(0.3),

    GRU(32),

    Dropout(0.3),

    Dense(
        16,
        activation="relu"
    ),

    Dense(
        1,
        activation="sigmoid"
    )
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        AUC(name="auc"),
        Precision(name="precision"),
        Recall(name="recall")
    ]
)

model.summary()
```
